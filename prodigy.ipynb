```python
#Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#Generate sample data
np.random.seed(42)
square_footage = np.random.rand(100,1) * 2000 +1000
num_bathrooms = np.random.randint(1,4, size=(100, 1))
num_bedrooms = np.random.randint(1,5, size=(100, 1))
house_prices = 100000 + 200 * square_footage + 5000 * num_bedrooms + 8000 *num_bathrooms +np.random.randn(100,1) * 10000

# Combine features into one matrix
X = np.hstack((square_footage, num_bedrooms, num_bathrooms))

#Add a column of ones to represent the intercept term
X_b = np.c_[np.ones((100,1)), X]

base = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(house_prices)
print("Intercept:", base[0])
print("Coefficients:", base[1:])

predict = X_b.dot(base)

#Plot the results
fig, ax = plt.subplots()
ax.scatter(square_footage, house_prices, color ='green' , label = 'Actual Price')
ax.scatter(square_footage, predict, color ='blue' , label = 'Predicted Price')
ax.set_ylabel('Square Footage')
ax.set_ylabel('House Price')
ax.legend()
plt.show()
```
